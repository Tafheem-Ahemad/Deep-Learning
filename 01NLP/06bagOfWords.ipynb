```python
import pandas as pd
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000, 
    formatter=dict(float=lambda x: "%.3g" % x))


mails=pd.read_csv('Datasets/spam classification/SMSSpamCollection.txt',
				sep="	",names=["name","messages"])
```

```python
mails
```

```python
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
lemmatizer=WordNetLemmatizer()
all_common_Words=stopwords.words('english')
```

```python
corpus=[]

for sen in mails['messages']:
	sen=sen[:]
	sen=sen.lower()
	sen=word_tokenize(sen)
	sen=[lemmatizer.lemmatize(word) for word in sen if word not in all_common_Words]
	sen=" ".join(sen)
	corpus.append(sen)
```

```python
corpus
```

##### min_df --> all vocabulary with count < min_dif will removed
##### max_df --> all vocabulary with count > max_dif will removed (may be spam)

```python
from sklearn.feature_extraction.text import CountVectorizer
CV=CountVectorizer(max_features=100,binary=True,min_df=0.005)
```

```python
# Not list of word , make a sentence and pass it
X=CV.fit_transform(corpus).toarray()

CV.vocabulary_
```

```python

X
```

```python
# N-Gram
from sklearn.feature_extraction.text import CountVectorizer
CV=CountVectorizer(max_features=50,ngram_range=(1,2))
```

```python
X=CV.fit_transform(corpus).toarray()

CV.vocabulary_
```

```python
X
```

```python
Y=mails['name']

from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
Y=encoder.fit_transform(mails['name'])
Y
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.25,random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()

model.fit(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score

print(classification_report(Y_test,Y_pred))
```

```python
print(accuracy_score(Y_test,Y_pred))
```
